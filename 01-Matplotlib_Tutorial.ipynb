## Thickness, Label and Legend

```python
fig=plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(data)
location=ax.set_xticks([0,10,20,30,40,50])
label=ax.set_xticklabels(
    ["one","two","three","four","five","six"],
    rotation=30,
    fontsize="small")

ax.set_title('First Graph')
ax.set_xlabel('Stages')
ax.set_ylabel('Anything')

feature={"title":"Perfect Graph", 
         "xlabel":"Steps", 
         "ylabel":"Numbers"}

ax.set(**feature)
```

## Adding Legends

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

data1=np.random.randn(1000).cumsum()
data2=np.random.randn(1000).cumsum()
data3=np.random.randn(1000).cumsum()

ax.plot(data1, 'k', label='one')
ax.plot(data2, 'b--', label='two')
ax.plot(data3, 'r.', label='three')

ax.legend()
plt.show()
```

## Annotations and Drawing on a Subplot

```python
import pandas as pd

df=pd.read_csv('D:/06_Data_Visualization/Datasets/FB.csv')
df.head()
```

```python
df = pd.read_csv('Datasets/FB.csv', index_col=0, parse_dates=True)
print(df.head())
```

```python
c_price=df["Close"]
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
c_price.plot(ax=ax, style='k-')
ax.set_title('Facebook stock closing prices')
c_price.min()
c_price["2018-12-24"]
date="2018-12-24"
text="Minimum"
ax.annotate(text, xy=(date,125), xytext=(date,150), arrowprops=dict(facecolor='red'))
```
